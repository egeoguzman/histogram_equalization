```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
#Find the all images in the folder and declare the path.
#Read the every images and add to the list.
imageformat=".jpg"
path="/Users/egeoguzman/Desktop/dip_hw2/inputs"
files=[os.path.join(path,f) for f in os.listdir(path)]
files
```

```python
#plt_func reads the images from the file and throw to the plot.hist function one by one
#img.ravel() merges the dimensions into a list and takes the every value as one dimension.
#[0,256] gives the interval of pixels
def plt_func(f):
    for i in f:
        img = cv.imread(i)
        plt.hist(img.ravel(),256,[0,256])
        plt.show()
plt_func(files)
```

```python
def hist_equ_func(f):
    for idx,i in enumerate(f):
        img = cv.imread(i,0)
        equalized_img = cv.equalizeHist(img)
        cv.imwrite('./hist_equ_output/output{}.png'.format(idx),equalized_img)
        plt.hist(equalized_img.ravel(),256,[0,256])
        plt.show()
hist_equ_func(files)
```
